# Question A1

Consider a linear inversion problem written in the form 

$$G\boldsymbol{m} = \boldsymbol{d}.$$

The following four mathematical expressions give possible solutions to this problem


1. $\boldsymbol{m} =  G^+\boldsymbol{d},$


2. $\boldsymbol{m} = G^T(GG^T)^{-1}\boldsymbol{d},$


3. $\boldsymbol{m} = G^{-1}\boldsymbol{d},$


4. $\boldsymbol{m} = (G^T G)^{-1} G^T  \boldsymbol{d}.$


For each of these approaches explain in words what type of inversion problem the approach can be used to solve, what properties the corresponding solution will possess, and any potential issues that might arise when attempting to utilise this solution approach.


 
 The following answers are based on [1]:
 
 
**1. Any type of inversion problem / Mixed-determined problems:**

The generalised inverse can be applied to provide a solution to any linear inverse ($G^+$ exists for any matrix).
 
**Properties of the solution:**

- If $G$ is square and non-singular, then $G^+=G^{-1}$, so that it is the regular inverse of $G$. (This is equivalent to case 3).

- If $G^TG$ is non-singular, then the generalised inverse defined above is equivalent to $G^+=(G^TG)^{-1}G^T$, so that it is the least-squares inverse of $G$. (This is equivalent to case 4: the columns are linearly independent - the left inverse).

- If $GG^T$ is non-singular, then $G^+=G^T(GG^T)^{-1}$, so that it is the minimum-norm inverse of $G$. (This is equivalent to case 2, where the rows are linearly independent - the right inverse).

As such:

- If there is any model parameter vector that can provide an exact fit to the data $\boldsymbol{d}$, then $\boldsymbol{m}^+$  will be such a vector. 

- If there is no model parameter vector that can provide an exact fit to the data $\boldsymbol{d}$, then $\boldsymbol{m}^+$ will provide a best least-squares fit to the data. 

- If there is more than one exact or best-fitting parameter vector to the data $\boldsymbol{d}$, then $\boldsymbol{m}^+$ will be the exact or best-fitting model that also has the minimal-norm. 


**Limitations:**

- The  generalised  inverse  only  solves  linear problems and most  real  inverse  problems  are  non-linear

- It is expensive to compute as the SVD calculation is expensive for large matrices. In practical terms it is impossible to use for the largest inverse problems.   

- No protection against very small singular values or very small eigenvalues that may be practically indistinguishable from zero when using finite-precision arithmetic. 

- If the ratio of the largest to the smallest singular value is very large, then the system will be very ill-conditioned.

- Small singular values will dominate the generalised inverse because they are reciprocated and become very large when calculating the inverse of $\Sigma$. Small singular values and small eigenvalues will  typically be dominated by noise.  The generalised inverse then will act to amplify noise in the data,


**2. Under-determined problems:**

It is used for fat $(m < n)$ matrices $A$ that are full rank - the matrix formed by $AA^T$  will not be singular.  

In under-determined problems, there are fewer equations than unknowns, and so  ð‘š  is less than  ð‘›. There are fewer data points than model parameters.

**Properties of the solution:**

- Since there are fewer equations than unknowns, this problem has a null space. As such, an infinite number of solutions will  satisfy the data equally well.    

- The minimum-norm inverse generates a perfect fit to the data.  The data is fitted and $\boldsymbol{m}^T\boldsymbol{m}$ is minimised.

**Limitations:**

- If  $G$  is  full  rank ($\text{rank}(G) = m$), then the problem is called purely under-determined. There will be infinitely many solutions that are able to fit the data exactly (as explained above).
 

**3. Equi-determined problems:**

It is used for general inverse problems, where the matrix A is square $(m = n)$ and has full rank if and only if it is not singular. 

The problem is to determine the model parameters $\boldsymbol{m}$ given a set of observational data $\boldsymbol{d}$: 

$$\boldsymbol{m} = G^{-1}(\boldsymbol{d}) $$

**Limitations:**

- The inverse equation has no exact solution at all (non-existence)

- Infinitely many solutions that all fit the data equally accurately (non-uniqueness)

- The solution for the model parameter  ð‘š  varies significantly when only very small changes are made to the data  ð‘‘  (instability)


**4. Over-determined problems:**

It is used for tall $(m > n)$ matrices $A$ that are full rank (the matrix formed by $A^TA$ will not be singular). In over-determined problems, there are more independent equations than unknowns.  

In these problems, there are  more  independent  data  points  than  there  are  model parameters. As such, the method presented is equivalent to performing a least squares curve fitting.

**Properties of the solution:**

- The inverse of  ðº  is not defined, and in general there is no solution for  ð‘š  that will exactly satisfy this relation.  As such, the problem is transformed to  $\boldsymbol{m} = (G^T G)^{-1} G^T  \boldsymbol{d}.$
 
- Its inverse will exist provided that  ðº  is of full column rank ( rank(ðº)=ð‘› ), that is provided that the problem is purely over-determined $G\, \boldsymbol{m} = \boldsymbol{d}$, $m>n$,  where $G$ is an $m\times n$ matrix.

**Limitations:**

- For $(G^TG)^{-1}$ to exist, the matrix $G$ must be of full column rank - its columns being linearly independent.

# Question A2

Consider the rectangular matrix

$$G = 
\begin{pmatrix}
2 & 2 & 10 & 10 \\
2 & 3 & 14 & 12 \\
4 & 5 & 24 & 22 
\end{pmatrix}.
$$

Convert this matrix into reduced-row-echelon form. 

What is the rank of this matrix? What dimensions are its range and its null space?

Find the appropriate number of vectors which can be used as a basis to span the null space.

Now consider the data vector
$$\boldsymbol{d} = \begin{pmatrix}
26 \\
30\\
56
\end{pmatrix}.$$


Use the pseudo-inverse (you may make use of `np.linalg.pinv`) to find a solution to the problem $G\boldsymbol{m}=\boldsymbol{d}$.

Demonstrate though some examples that adding any multiples of the vectors spanning the null space to this solution (i.e. adding any vector from within the null space) also provides a valid solution to this problem.

Demonstrate that the solution obtained using the pseudo-inverse is the minimal norm solution to this problem.


The rectangular matrix:
    
$$G = 
\begin{pmatrix}
2 & 2 & 10 & 10 \\
2 & 3 & 14 & 12 \\
4 & 5 & 24 & 22 
\end{pmatrix}.
$$

Can be expressed in RREF as: 

$$G = 
\begin{pmatrix}
1 & 0 & 1 & 3 \\
0 & 1 & 4 & 2 \\
0 & 0 & 0 & 0 
\end{pmatrix}.
$$

The rank of the matrix is 2, as it can be seen from the number of non-zero rows in the RREF form. There are two linearly independent row vectors.

The range of the matrix is the dimension of its column space. As it can be seen from the RREF form, the first 2 columns are linearly independent. The 3rd and 4th columns are a linear combination of the first two. As such, the range of the matrix is the 2-dimensional space spanned by:

$$\boldsymbol{c_1} = \begin{pmatrix}
2 \\
2\\
4
\end{pmatrix}.$$

$$\boldsymbol{c_2} = \begin{pmatrix}
2 \\
3\\
5
\end{pmatrix}.$$


The number of columns of a matrix is equal to the sum of its rank and nullity (the dimension of the null space). Since the rank is 2, the nullity must be 2. As such, 2 vectors are used as a basis to span the null space. This can be checked, and the vectors that span the null space can be calculated, using the following code snippet: 

```python
from sympy import Matrix 
A = [[2, 2, 10, 10], [2, 3, 14, 12], [4, 5, 24, 22]]  
A = Matrix(A) 
NullSpace = A.nullspace()   # Here NullSpace is a list  
NullSpace = Matrix(NullSpace)   # Here NullSpace is a Matrix 
print("Null Space : ", NullSpace) 
NoC = A.shape[1] 
# Rank of A 
rank = A.rank() 
# Nullity of the Matrix 
nullity = NoC - rank  
print("Nullity : ", nullity) 
```


The null space is spanned by the 2 following vectors:
$$\boldsymbol{v_1} = \begin{pmatrix}
-1 \\
-4\\
1\\
0
\end{pmatrix}.$$

$$\boldsymbol{v_2} = \begin{pmatrix}
-3 \\
-2\\
0\\
1
\end{pmatrix}.$$

And it can be checked by:

```python
G = np.array([[2, 2, 10, 10], [2, 3, 14, 12], [4, 5, 24, 22]])
v1 = np.array([-1, -4, 1, 0])
v2 = np.array([-3, -2, 0, 1]) 
print(np.dot(G, v1))
print(np.dot(G, v2))
```


The problem $G\boldsymbol{m}=\boldsymbol{d} $  can be expressed as $\boldsymbol{m} =  G^+\boldsymbol{d} $ by making use of the pseudo-inverse. This is calculated in the following code snippet:

```python
import numpy as np
d = np.array([26, 30, 56])
Gp = np.linalg.pinv(G)
print(Gp)
m = np.dot(Gp, d)
print(m)
```


Adding any multiples of the vectors spanning the null space to this solution (i.e. adding any vector from within the null space) also provides a valid solution to this problem.

This means that $G\boldsymbol{(m + a*v_1)}=\boldsymbol{d} $ and $G\boldsymbol{(m + b*v_2)}=\boldsymbol{d} $, where a and b are any integer.

This is demonstrated in the following code snippet:

```python
print(np.isclose(np.dot(G, m), d)) # first let's check that the solution obtained does indeed satisfy the system
print(np.isclose(np.dot(G, m+v1), d)) # first let's check that the solution obtained does indeed satisfy the system
print(np.isclose(np.dot(G, m+v2), d))
print(np.isclose(np.dot(G, m+2*v1), d)) # first let's check that the solution obtained does indeed satisfy the system
print(np.isclose(np.dot(G, m+2*v2), d))
print(np.isclose(np.dot(G, m+247*v1), d)) # first let's check that the solution obtained does indeed satisfy the system
print(np.isclose(np.dot(G, m+29*v2), d))
```


The code below shows that the solution obtained using the pseudo-inverse is the minimal norm solution to this problem.    

```python
from numpy import linalg as nla

# is x_m the minimum norm solution?

# plot the norm of the vectors we get by adding on multiples of the numm vector
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)

mult = np.linspace(-10,10,100)
norms = []
for mu in mult:
    print
    norms.append(nla.norm(m+mu*v1))

ax1.plot(mult,norms)

mult = np.linspace(-10,10,100)
norms = []
for mu in mult:
    print
    norms.append(nla.norm(m+mu*v2))

plt.title('Norm of solution as a function of multiples of null vectors')
plt.xlabel('Multiple')
plt.ylabel('Norm')
ax1.plot(mult,norms)
```


The plot above shows that the minimal norm solution is achieved if the vectors that span the null space are not added to it.

```python
# PIL = Python Imaging Library
from PIL import Image
img_colour = Image.open('./Imperial_College_Royal_School_of_Mines.jpg')
# convert to grey scale
img_grey = img_colour.convert('LA')

fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
ax1.imshow(img_colour)

# use getdata, first band - https://www.geeksforgeeks.org/python-pil-image-getdata/
img_mat = np.array(img_grey.getdata(band=0), float)
# turn into matrix of the correct shape
img_mat.shape = (img_grey.size[1], img_grey.size[0])

fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
#plot using a gray colormap
ax1.imshow(img_mat, cmap='gray')

img_colour = Image.open('./Flag_of_Spain.jpg')
# convert to grey scale
img_grey = img_colour.convert('LA')
# what does LA mean? ... https://pillow.readthedocs.io/en/stable/handbook/concepts.html#concept-modes

fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
ax1.imshow(img_colour)

# use getdata, first band - https://www.geeksforgeeks.org/python-pil-image-getdata/
img_mat1 = np.array(img_grey.getdata(band=0), float)
# turn into matrix of the correct shape
img_mat1.shape = (img_grey.size[1], img_grey.size[0])
fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
#plot using a gray colormap
ax1.imshow(img_mat1, cmap='gray')

img_colour = Image.open('./rowing_machine.jpg')
# convert to grey scale
img_grey = img_colour.convert('LA')
# what does LA mean? ... https://pillow.readthedocs.io/en/stable/handbook/concepts.html#concept-modes

fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
ax1.imshow(img_colour)

# use getdata, first band - https://www.geeksforgeeks.org/python-pil-image-getdata/
img_mat2 = np.array(img_grey.getdata(band=0), float)
# turn into matrix of the correct shape
img_mat2.shape = (img_grey.size[1], img_grey.size[0])

fig = plt.figure(figsize=(8, 12))
ax1 = fig.add_subplot(111)
#plot using a gray colormap
ax1.imshow(img_mat2, cmap='gray')

# find the SVD
U, sigma, VT = np.linalg.svd(img_mat)
plt.figure()
x = np.linspace(0, sigma.size-1, num = sigma.size)
sigma_sum = sum(sigma)
ratio = np.zeros(sigma.size)
for i in range(1,sigma.size+1):
    # restrict the number of singular values considered
    sigma_r = sigma[:i]
    sigma_r_sum = sum(sigma_r)
    ratio[i-1] = sigma_r_sum/sigma_sum
plt.plot(x, ratio, label = "Royal School of Mines")

U, sigma, VT = np.linalg.svd(img_mat1)
x = np.linspace(0, sigma.size-1, num = sigma.size)
sigma_sum = sum(sigma)
ratio = np.zeros(sigma.size)
for i in range(1,sigma.size+1):
    # restrict the number of singular values considered
    sigma_r = sigma[:i]
    sigma_r_sum = sum(sigma_r)
    ratio[i-1] = sigma_r_sum/sigma_sum
plt.plot(x, ratio, label = "Spanish flag")

U, sigma, VT = np.linalg.svd(img_mat2)
x = np.linspace(0, sigma.size-1, num = sigma.size)
sigma_sum = sum(sigma)
ratio = np.zeros(sigma.size)
for i in range(1,sigma.size+1):
    # restrict the number of singular values considered
    sigma_r = sigma[:i]
    sigma_r_sum = sum(sigma_r)
    ratio[i-1] = sigma_r_sum/sigma_sum
plt.title('Cumulative sum of singular values as a fraction of sum of all singular values ')
plt.xlabel('Number of singular values')
plt.ylabel('Cumulative sum of singular values / \nsum of all singular values')
plt.plot(x, ratio, label = "Rowing machine")
plt.legend()
```


    
The number of singular values obtained after performing an SVD decomposition contains the full rank information of an image [2]. To compress an image, some of these singular values can be neglected to perform a low-rank approximation. 

The magnitude of singular values is related to that of their corresponding eigenvalues.  Eigenvalues describe the magnitude of change in the direction of their eigenvectors [3]. The larger a singular value, the larger the magnitude of the change, and therefore, the more information it contains about the image. Performing SVD results in an arrangement of singular values in descending order. The first singular value contains the most significant amount of information, and subsequent singular values contain decreasing amounts of image information [4]. The first singular value of the flag picture contains over 50% of the information of the image, whereas that of the RSM picture only contains above 10%. 

Computing a ratio of the first $n$ singular values as a fraction of the sum of all the singular values is the same as calculating a ratio of the cumulative magnitude of change over the total magnitude of change. As this ratio approaches 1, the singular values approach minimal values. Such small values do not contain much information about the image. The quicker the ratio approaches 1, the better the quality of a low-rank approximation -- less singular values are needed to compress or recover the image. 

As such, this ratio is a useful metric to determine how well the information in the image is represented by the low-rank approximation. Similarly, the simpler the image, the lesser the number of singular values needed to describe it and the lower the rank of the matrix that describes it [5]. As a result, an image with a low number of singular values (such as a flag) would be easier to compress than one with many singular values (such as the RSM picture). 

When comparing a complex image (Royal School of Mines) with an image of medium complexity (rowing machine) and with a simple image (the Spanish flag), a difference in the rate of convergence to 1 can be observed. As expected, the simpler the image, the quicker the ratio converges to 1, and the lesser the number of singular values obtained from the SVD.
