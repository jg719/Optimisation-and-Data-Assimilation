The forced one-dimensional viscous Burgers' equation occurs frequently in fluid dynamics and acoustics. In form it is very similar to the Navier-Stokes equations you saw in ACSE-3,

$$ \frac{\partial u}{\partial t} +u\frac{\partial u}{\partial z} = \nu \frac{\partial^2 u}{\partial z^2} +r(z).$$

Here $u$ is the variable of interest, $z$ is the independent variable representing space in the domain and $r(z)$ is a given forcing term. For numerical convenience it is also common to write the advective term $u\frac{\partial u}{\partial z}$ in a conservative form,
$$ \frac{\partial u}{\partial t} +\frac{1}{2}\frac{\partial u^2}{\partial z} = \nu \frac{\partial^2 u}{\partial z^2} +r(z).$$

You have been provided with observational data of $u$ for a number of time levels at $z=5$ and $z=-5$ from a Burger's model system in a periodic domain of period 20. In the region $[-10, 10]$ $r(z)$ takes the form $100\sin\left(\frac{3t}{8}\right)\exp(-100z^2)$.

To decide between single/multiple data assimilation strategies, an assumption for the units of the time window of the forecast (t = 100) is needed [1]. If the order of time is of days, multiple assimilation strategies might be preferable in order to account for key seasonal changes that impact the background state vector. 2.	**Multiple 4D-Var Assimilations:** Similar to the Single 4D-Var Assimilation, this method is more accurate than Multiple Kalman Filter Assimilations and Multiple 3D-Var Assimilations. As such, very slight improvements in accuracy might come at a tremendous computational expense.

While the Kalman Filter takes into consideratiion the time dependence (at a high computational expense), it does not account for non-linear observations (as does 3D-Var). 3D-Var could be more appropriate if the time window is small, the observations are very non-linear or the computing power is constrained. Using Multiple Kalman Filter Assimilations poses the question of whether it is worth computing very accurate iterations for a method that is unaccurate in most real non-linear problems.   


4.	**Multiple 3D-Var Assimilations:** For linear problems, this method is equivalent to Multiple OI Assimilations. In the case of non-linear observations, 3D-var provides a more accurate result than single or multiple OI. Performing Multiple 3D-Var Assimilations might result in very subtle increases in accuracy at the expense of a significant increase in computational overhead.


5.	**Single OI Assimilation:** This method is the cheapest, but it is also the least accurate. It treats all the observations as of equal relevance [7], giving the same weight to an observation from yesterday than to one taken a month ago. As a result, a single OI assimilation might be preferred over multiple OI, priming computing efficiency over a marginal accuracy increase.

## Single 4D-Var Data Assimilation: Initial settings and parameters

```python
%matplotlib inline
from scipy.optimize import minimize
import scipy.optimize as sop
from scipy.optimize.linesearch import line_search_armijo as line_search
import scipy.linalg as sl
from scipy.linalg import inv
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.style.use('seaborn-pastel')
```

```python
# observations data is in array form y[timelevel, location]
# where location 0 is z=-5 and location 1 is z=5
# The actual time value is stored in the t_obs array
data = np.load('burgers_observations.npz')
t_obs = data['t_obs']
y = data['y']

nt = 79
#nx = 100
z = np.linspace(-10, 10, 61)
nz = len(z)
dz = z[1] - z[0]
u = np.full(z.shape, 1)
nu = 0.4
dt = 0.05
x_b = np.ones(61) # define background state -- all zeros
```

```python
# B is the (sparse) background error covariance matrix
def generate_B_matrix(z):
    B = np.empty([len(z), len(z)])
    
    for _ in range(len(z)):
        B[_, :] = 0.2**2*np.exp(-4*(z[_]-z)**2)*((z[_]-z)**2<=4)
    return B


B = generate_B_matrix(z)
Binv = np.linalg.inv(B)
```

## Derivation of the observation operator $H$

<div class="optional">
The aim of the observation operator is to take observations at any space location and time, and interpolate them into the spatial and temporal discretisation used in the model. 
 
 The observations provided correspond to locations $z = -5$ and $z = 5$ and time levels $t = [0, 79]$. As such, the observation operator $H$ must have as many rows as observation positions and as many columns as observation times (in this case, 2 rows x 80 columns). 
 
 For simplicity, it is assummed that the levels of the time discretisation match the time levels of the observations provided. As such, $dt$ must be selected such that the modulus between $1$ and $dt$ is $0$. This avoids the need to interpolate in time. 
 
A simple linear function is used to interpolate in space. This is implemented in the code below (works for different $dz$). The space discretisation used, however, matches the space locations of the observations. Again, this further simplifies coding the operator. 

```python
# Code H
h = np.zeros((nt, 2, len(z)))
h_x_pos_0 = int(1/4*nz) # position of 1st observation in h 
h_x_pos_1 = int(3/4*nz) # position of 2nd observation in h

# position of observations
x1 = 5 # corresponding to position -5
x2 = 15 # corresponding to position 5
# calculate weights for first observation
h[:, 0, h_x_pos_0] = (abs((x1 - h_x_pos_0*dz)/dz))
h[:, 0, h_x_pos_0 + 1] = (1-abs((x1 - h_x_pos_0*dz)/dz))

# calculate weights for second observation
h[:, 1, h_x_pos_1] = (abs((x2  - h_x_pos_1*dz)/dz))
h[:, 1, h_x_pos_1 + 1] = (1-abs((x2 - h_x_pos_1*dz)/dz))
```

```python
## For the problem provided, the observation operator can also be simply described as:
h = np.zeros((nt, 2, 61))
h[:, 0, 15] = 1
h[:, 1, 45] = 1
```

## Derivation of the observation error covariance matrix $R$

<div class="optional">
The observations  have Gaussian errors with standard deviation 0.05 (at  𝑧=−5  ) and 0.1 (at  𝑧=5 ) and are independent (indicating that $R$ must be diagonal). 


The error covariance matrix $R$ is given by:

$$ R = \begin{pmatrix}  
\sigma_{0,0}^2 & 0 \\
0 & \sigma_{1, 1}^2 \\
\end{pmatrix} $$


```python
# Code R
R = np.zeros((2, 2))
R[0, 0] = 0.05**2 #calculate variance of 1st observation
R[1, 1] = 0.1**2 #calculate variance of 2nd observation
Rinv = np.linalg.inv(R)
```

## Forward model (discretised Burger's equation, non-linear)

```python
## Use forward Euler for the (nonlinear) advection  and Crank-Nicolson for diffusion
def rhs(U, dt, t):
    out = np.empty(U.shape)
    out[0] = U[0]-0.5*dt/dz*(U[1]**2-U[-1]**2-nu*(U[-1]-2*U[0]+U[1])/dz)
    out[1:-1] = U[1:-1]-0.5*dt/dz*(U[2:]**2-U[:-2]**2-nu*(U[:-2]-2*U[1:-1]+U[2:])/dz)
    out[-1] = U[-1]-dt/dz*(U[0]**2-U[-2]**2-nu*(U[-2]-2*U[-1]+U[0])/dz)
    
    return out+dt*10*np.sin(.375*np.pi*(t+0.5*dt))*np.exp(-100*z**2)

def lhs_matrix(u, dt):
    mat = np.eye(u.shape[0])
    mat[0,-1] -= 0.5*dt*nu/dz**2
    mat[0, 0] += dt*nu/dz**2
    mat[0, 1] -= 0.5*dt*nu/dz**2
    for _ in range(1, u.shape[0]-1):
        mat[_, _-1] -= 0.5*dt*nu/dz**2
        mat[_, _] += dt*nu/dz**2
        mat[_, _+1] -= 0.5*dt*nu/dz**2
    mat[-1,-2] -= 0.5*dt*nu/dz**2
    mat[-1, -1] += dt*nu/dz**2
    mat[-1, 0] -= 0.5*dt*nu/dz**2
    return mat

def forward_model(u0, t = 0.0, t_final = 79):
    dt = 0.05
    u = u0
    out = [u0]
    mat = lhs_matrix(u0, dt)
    matinv = inv(mat)
    while t<t_final:
        for _ in range(10):
            u = matinv.dot(rhs(u, dt, t))
            t += dt
        out.append(u)
    out = out[0::2]
    return out
```

## Cost function and gradient of cost function

<div class="optional">
The cost function in 4D-var is given by:

$$ \mathcal{J}(\mathbf{x_0}) = \frac{1}{2}\left[(\mathbf{x}_b-\mathbf{x_0})^T\mathbf{B}^{-1}(\mathbf{x}_b-\mathbf{x}_0)+\sum_i(\mathbf{y}-h_i(\mathbf{x}(t_i))))^T\mathbf{R}_i^{-1}(\mathbf{y}-h_i(\mathbf{x}(t_i)))\right], $$

subject to the constraints that 
$$\mathbf{x}^{(0)} = \mathbf{x}_0, $$
$$ \mathbf{x}^{(k+1)} = \mathbf{m}^{(k)}\left(\mathbf{x}^{(k)}\right),$$
$$x(t) = f(t, \mathbf{x}^{(0)}, \mathbf {x}^{(1)}, \ldots, ),$$
for some temporal interpolation function, $f$.

Using a set of Lagrange multipliers leads to the following equation set:

$$ \nabla_{\mathbf{x}_0} \mathcal{J} = \mathbf{B}^{-1}(\mathbf{x}_b-\mathbf{x}_0)+\lambda_0$$
$$ \mathbf{x}^{(k+1)} = \mathbf{m}^{(k)}\left(\mathbf{x}^{(k)}\right) $$
$$ \mathbf{\lambda}_i = \mathbf{M}^T\mathbf{\lambda}_{i+1} +\sum_j \frac{\partial f_j}{\partial \mathbf{x}_i}H_j^T\mathbf{R}^{-1}(\mathbf{y}-\mathbf{h}(\mathbf{x}(t_j)))$$
$$ \lambda_N = \sum_j \frac{\partial f_j}{\partial \mathbf{x}_N}H_j^T\mathbf{R}^{-1}(\mathbf{y}-\mathbf{h}(\mathbf{x}(t_j)))$$

These are implemented below.

```python
def j(x0): # define cost function
    x = forward_model(x0)
    j_b0 = (x0-x_b)@(Binv@(x0-x_b)) 
    j_b = j_b0
    for t in range(nt):
        j_b += (y[t, :]-h[t, ...]@(x[t])).T@(Rinv.dot(y[t, :]-h[t, ...]@(x[t])))
    return 1/2*j_b

#define derivative of cost function
def jac(x0): # define derivative of cost function 
    x = forward_model(x0)
    jac= Binv.dot(x0-x_b) + adjoint_model(x, y, h)
    return jac #change this back

def jac_taylor_test(x0): # wrapper to sum the result of the jacobian (to pass to the Taylor test later on) 
    return sum(jac(x0))
```

## Derivation of the adjoint model $M^t$

<div class="optional">
The inviscid Burger's equation is given by:

$$\frac{\partial u}{\partial t} +u\frac{\partial u}{\partial z} = \nu \frac{\partial^2 u}{\partial z^2} $$

It is a non-linear equation. As such, the aim is to linearise it and discretise it to arrive to the adjoint model needed to compute the gradient of the cost function. 

To linearise it, we use the tangent linear (TL) method. A perturbation $ u = u + \delta u$ is introduced, leading to the tangent linear version of Burger's equation:

<div class="optional">
$$ \frac{\partial \delta u}{\partial t} +u\frac{\partial \delta u}{\partial z} + \delta u \frac{\partial u}{\partial z} - \nu \frac{\partial^2 \delta u}{\partial z^2} = 0 $$

This equation can be discretised in space (central differences) and time (forward differences):

$$ \delta u_1^i = \delta u_0^i + \Delta t(- u_0^i \frac{\delta u_0^{i+1} -  \delta u_0^{i-1}}{2 \Delta z} - \delta u_0^i \frac{u_0^{i+1}-u_0^{i-1}}{2\Delta z} + \nu \frac{\delta u_0^{i+1}-\delta 2u_0^i+\delta u_0^{i-1}}{(\Delta z)^2})$$

The expression above corresponds to an implicit system of equations that can be expressed as a matrix (the forecast model matrix M):


<div class="optional">
$$ M = \begin{pmatrix}  
  1 + \Delta t(\frac{u_0^{i+1}- u_0^{i-1}}{2\Delta z} - \frac{2 \nu}{(\Delta z)^2})  & \Delta t(\frac{-u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 0 & 0 & ... 
\\
\Delta t(\frac{u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 1 + \Delta t(\frac{u_0^{i+1} - u_0^{i-1}}{2\Delta z} - \frac{2 \nu}{(\Delta z)^2}) & \Delta t(\frac{-u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 0 & ...
\\
0 &  \Delta t(\frac{u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 1 + \Delta t(\frac{u_0^{i+1} - u_0^{i-1}}{2\Delta z} - \frac{2 \nu}{(\Delta z)^2}) &  \Delta t(\frac{-u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & ...\\
\\
... & ... & ... & ... & ... \\
\end{pmatrix} $$



<div class="optional">
In parallel, the adjoint equation is derived as a sanity check for the matrix above. Applying Lagrange multipliers to the TL equation leads to: 

$$\int \lambda (\frac{\partial \delta u}{\partial t} +u\frac{\partial \delta u}{\partial z} + \delta u \frac{\partial u}{\partial z} - \nu \frac{\partial^2 \delta u}{\partial z^2})   d\Omega dt = 0$$


<div class="optional">
Integrating by parts, this leads to the continuos adjoint equation: 

$$ \frac{\partial \lambda}{\partial t} + u\frac{\partial \lambda}{\partial z} - \lambda \frac{\partial u}{\partial z} + \nu \frac{\partial^2 \lambda}{\partial z^2} = 0  $$


<div class="optional">
Discretising in space and time leads to the following expression:
$$ \lambda_1^i = \lambda_0^i + \Delta t(\nu \frac{\lambda_0^{i+1}- 2\lambda_0^{i} + \lambda_0^{i-1}}{(\Delta z^2)} - u_0^i \frac{\lambda_0^{i+1} - \lambda_0^{i-1}}{2\Delta z} + \lambda_0^{i} \frac{u_0^{i+1} - u_0^{i-1}}{2\Delta z})$$ 

<div class="optional">
As such the discrete adjoint equation can be used to determine $\lambda_0$ backwards, starting from $\lambda_N$. To calculate $\lambda$ backwards, an implicit scheme such as Backward Time Centred Space (BTCS) must be used. This requires the problem to be expressed as a system of linear equations, leading to the same matrix M indicated above. 
Having checked the correctness of $M$, the adjoint model $M^t$ can be derived:

$$ M^t = \begin{pmatrix}  
  1 + \Delta t(\frac{u_0^{i+1}- u_0^{i-1}}{2\Delta z} - \frac{2 \nu}{(\Delta z)^2})  & \Delta t(\frac{u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 0 & 0 & ... 
\\
\Delta t(\frac{-u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 1 + \Delta t(\frac{u_0^{i+1} - u_0^{i-1}}{2\Delta z} - \frac{2 \nu}{(\Delta z)^2}) & \Delta t(\frac{u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 0 & ...
\\
0 &  \Delta t(\frac{-u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & 1 + \Delta t(\frac{u_0^{i+1} - u_0^{i-1}}{2\Delta z} - \frac{2 \nu}{(\Delta z)^2}) &  \Delta t(\frac{u_0^i}{2\Delta z} + \frac{\nu}{(\Delta z)^2}) & ...\\
\\
... & ... & ... & ... & ... \\
\end{pmatrix} $$



<div class="optional">
It is important to note that $ \lambda_N $ can be calculated, as given by:
$$ \lambda_N = \sum_j \frac{\partial f_j}{\partial \mathbf{x}_N}H_j^T\mathbf{R}^{-1}(\mathbf{y}-\mathbf{h}(\mathbf{x}(t_j)))$$


```python
def adjoint_model(u0, y, h, nt = 80):
    u = np.zeros((nt, nz))
    u[:, :] = u0 # set initial condition from forward model
    l = -h[0, ...].T.dot(Rinv.dot(y[-1, :]-h[0, ...].dot(u[-1]))) # calculate lambda_n
    for t in range(1, nt+1): #update M for each time step, integrating backwards in time (-t)
        M = np.zeros((nz, nz))
        for i in range(0, nz-1): # build M 
            M[i, i] = 1+dt*((u[-t, i+1]-u[-t, i-1])/(2*dz) - 2*nu/(dz**2))
            M[i, i+1] = dt*(-u[-t, i]/(2*dz)+nu/(dz**2))
            M[i+1, i] = dt*(u[-t, i]/(2*dz)+nu/(dz**2))
        M[-1, -1] = 1+dt*((u[-t,0]-u[-t, -2])/(2*dz) - 2*nu/(dz**2)) #don't forget last diagonal term in M
        l =M.T@l -(h[0, ...].T)@Rinv@(y[-t, :]-h[0, ...]@(u[-t])) + dt*10*np.sin(.375*np.pi*(t+0.5*dt))*np.exp(-100*z**2)# calculate lambda_n_1
    return l  
```

```python
plt.figure(figsize=(16, 6))
x = forward_model(np.ones(61))
plt.plot(z, adjoint_model(x, y, h))
plt.title('$\lambda_0$ (as derived from adjoint)')
plt.ylabel('$\lambda_0$')
plt.xlabel('$z$')
```

## Taylor test: Validation of the gradient of the cost function

```python
def taylor_test(f, grad_f, x, h0, max_iter=20, plot_convergence=True, print_convergence=True):
    """Taylor test to verify that the function grad_f is the derivative of the function `f`
    
    We test:
    
       f(x + h) = f(x) + grad_f(x).h + O(h^2)
    
    f, grad_f  - function and its derivative to test
    x          - point in which to test
    h0         - initial perturbation
    max_iter   - number of times that h is halved
    plot_convergence - whether to plot the convergence of the Taylor residual f(x+h)-f(x)-grad_f(x).h
    print_convergence - prints the order of convergence between subsequent iterations."""


    h = h0[0]
    residuals = []
    hnorms = []
    for i in range(max_iter):
        r = f(x + h) - f(x) - grad_f(x) * h
        residuals.append(sl.norm(r))
        hnorms.append(sl.norm(h))
        h /= 2.

    if plot_convergence:
        fig, ax = plt.subplots(1,2, figsize=(16,4))
        ax[0].semilogy(residuals)
        ax[0].set_xlabel('iteration')
        ax[0].set_ylabel('Taylor residual $|r|$')
        ax[0].set_xticks(range(0,max_iter,2))
        ax[1].loglog(hnorms, residuals)
        ax[1].set_xlabel('perturbation size $\|h\|$')
        ax[1].set_ylabel('Taylor residual $|r|$')##
    if print_convergence:
        residuals = np.array(residuals)
        print('Order of convergence, log(r(h_i)/r(h_{i+1}))/log(2):')
        print(np.log(residuals[:-1]/residuals[1:])/np.log(2))
```

```python
np.random.seed(0)  
h0 = np.random.random(1)/10 #set original perturbation
taylor_test(j, jac, x_b, h0)
```

<div class="optional">
The order of convergence, computed from subsequent iterations, rapidly reaches the expected value of 1.

## Optimisation of the cost function to find an optimal initial conditions ($x_0$ or $x_b$)

A local minimum of the cost function can be found through the use of an optimisation algorithm. In 1D, line search is the most common option. 

```python
# optimisation via Scipy's LBFGS
res = minimize(j, np.ones(61), method= 'L-BFGS-B'  , jac=jac, tol  = 1e-2)
x_a_bfgs = res['x']
```

```python
# optimisation via Scipy's CG
res = minimize(j, x_b, method='CG', jac = jac, tol = 1e-2)
x_a_cg = res['x']
```

```python
# optimisation via line search
x_a = 1*np.ones(61) 
j0 = j(x_a)
for _ in range(100):
    j0 = j(x_a)
    pk = -jac(x_a) # define optimisation bounds
    res = line_search(j, x_a, pk, -pk, j0, alpha0=0.01)  # a step magnitude in the search direction of a=0.01 works well in practice
    print("Iteration: {}".format(_))
    print(res, "\n")
    x_a += res[0]*pk
    j0 = res[2]
    if res[0]<1e-7: 
        break
```

```python
plt.figure(figsize=(16, 6))
plt.plot(z, np.ones(len(z)), label = 'Initial background' )
plt.plot(z, x_a, label = 'Line Search: New $x_b$ (or $x_0$) / Final $x_a$')
plt.plot(z, x_a_bfgs, label = 'BFGS: New $x_b$ (or $x_0$) / Final $x_a$')
plt.plot(z, x_a_cg, label = 'CG: New $x_b$ (or $x_0$) / Final $x_a$')
plt.title('Optimal $z_0$ (or $z_b$) after optimisation / DA ($z_b = 1$)')
plt.ylabel('$z_0(z)$')
plt.xlabel('$z$')
plt.legend()
```

## Comparison between observations, initial model and model after DA at $z=-5$ and $z=5$ as a function of time

```python
forecast_15 = []
forecast_45 = []
original_15 = []
original_45 = []

x_t = forward_model(np.ones(nz), t_final = 100) # forecast before DA
x_f = forward_model(x_a, t_final = 100) # forecast after DA

t = np.linspace(0, 79, 80) # time axis

error = np.sqrt(B[15, 15]) # calculate standard deviation from background error covariance matrix

for i in range(80):
    original_15.append(x_t[i][15])
    forecast_15.append(x_f[i][15])
    
forecast_15 = np.array(forecast_15)

plt.figure(figsize=(16, 4))
plt.plot(t, forecast_15, color = 'orange', label = 'Forecast after DA')
plt.errorbar(t, y[:, 0], yerr=2*0.05, fmt = 'o', color = 'silver', label = 'Observations');
plt.fill_between(t, forecast_15-error, forecast_15+error, facecolor='#ffeead')
plt.title('Forecast after DA and observations (z = -5, t = [0, 79])')
plt.ylabel('u(z, t)')
plt.xlabel('t')
plt.legend()

error = np.sqrt(B[45, 45]) # calculate standard deviation from background error covariance matrix

for i in range(80):
    original_45.append(x_t[i][45])
    forecast_45.append(x_f[i][45])


forecast_45 = np.array(forecast_45);

plt.figure(figsize=(16, 4))
plt.plot(t, forecast_45, color = 'orange',  label = 'Forecast after DA')
plt.fill_between(t, forecast_45-error, forecast_45+error, facecolor='#ffeead')
plt.errorbar(t, y[:, 1], yerr=2*0.1, fmt = 'o', color = 'silver', label = 'Observations')
plt.title('Forecast after DA and observations (z = 5, t = [0, 79])')
plt.ylabel('u(z, t)')
plt.xlabel('t')
plt.legend()
```

<div class="optional">
The error bounds are provided by the background error covariance matrix. The diagonal of such a covariance matrix contains the "autocovariances" of each of the space levels being optimised. The aim is to calculate the standard deviation corresponding to entries -5 and 5. To do so, the square root of the diagonal entry of the background error covariance matrix that corresponds to these locations is taken.

$$\sigma_{1}(z=-5) = \sigma_{2}(z=5) = \sqrt{\sigma^2} = \sqrt{0.04} = 0.2$$

## Analysis of residual errors (original and forecast after DA)

```python
original_distance = [] #stored normalised absolute distance between original forecast and observations
forecast_distance = [] #stored normalised absolute distance between forecast after DA and observations
res_f_15 = [] #residuals original forecast and observations at z=-5
res_f_45 = [] #residuals forecast after DA and observations at z=-5
res_t_15 = [] #residuals original forecast and observations at z=5
res_t_45 = [] #residuals original forecast and observations at z=5
for i in range(80):
    original_distance.append(abs(x_t[i][15]- y[i][0]) + abs(x_t[i][45] - y[i][1])/len(z))
    forecast_distance.append(abs(x_f[i][15]- y[i][0]) + abs(x_f[i][45] - y[i][1])/len(z))
    res_f_15.append(x_f[i][15]- y[i][0])
    res_f_45.append(x_f[i][15]- y[i][1])
    res_t_15.append(x_t[i][15]- y[i][0])
    res_t_45.append(x_t[i][15]- y[i][1])
#normalise residuals by dividing by maximum
res_f_15 = res_f_15/max(res_f_15)
res_f_45 = res_f_45/max(res_f_45)
res_t_15 = res_t_15/max(res_t_15)
res_t_45 = res_t_45/max(res_f_45)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (16, 3))
ax1.hist(res_t_15, bins = 10)
ax2.hist(res_f_15, bins = 10)
ax1.title.set_text('Distribution of residuals (original forecast, z=-5)')
ax2.title.set_text('Distribution of residuals (forecast after DA, z=-5)')
ax1.set_xlabel('Residual error')
ax2.set_xlabel('Residual error')
ax1.set_ylabel('Frequency')
ax2.set_ylabel('Frequency')

fig, (ax3, ax4) = plt.subplots(1, 2, figsize = (16, 3))
ax3.hist(res_t_45, bins = 10)
ax4.hist(res_f_45, bins = 10)
ax3.title.set_text('Distribution of residuals (original forecast, z=5)')
ax4.title.set_text('Distribution of residuals (forecast after DA, z=5)')
ax3.set_xlabel('Residual error')
ax4.set_xlabel('Residual error')
ax3.set_ylabel('Frequency')
ax4.set_ylabel('Frequency')
```

<div class="optional">
The distribution of the residual errors is expected to be normally distributed around zero. As it can be seen from the difference between the plots on the left and those on the right, the forecast after DA seems to be distributed closer to the expected "bell" (i.e., Gaussian) shape. In particular, the distribution of residuals after performing DA at $z = -5$ appears to be unbiased, whereas that at $z = 5$ is clearly biased (shifted and skewed to the left with respect to zero). This shows that the current forecast tends to underestimate the provided observations at the position $z = 5$.

## Extension: Single, multiple OI (window length = 10) and Kalman filter

```python
#### Single OI
x_b_2 = np.ones(len(z))
x_a_2 = np.ones(len(x_b))
z_n = np.zeros(2)
for i in range(len(y)):
    W = B.dot((h[0, ...].T)).dot(inv(R+h[0, ...].dot(B.dot(h[0, ...].T))))
    z_n = y[i, :] - h[0, ...].dot(x_b_2)
    x_a_2 = x_b_2 + W.dot(z_n.T)

x_b = np.ones(len(z)) # set initial background
x_a = np.ones(len(x_b)) # set initial analysis
z_n = np.zeros(2)
window_length = 10; #window length set to 10
count = 1 

#### Multiple OI
while count<9 :
    for i in range(window_length*(count-1), window_length*(count)):    
        W = B.dot((h[0, ...].T)).dot(inv(R+h[0, ...].dot(B.dot(h[0, ...].T))))
        z_n = y[i, :] - h[0, ...].dot(x_b)
        x_a = x_b + W.dot(z_n.T)
    x_b = x_a #reset background vector after going through each OI assimiliation window
    count += 1       
    

#### Kalman Filter attempt   
def construct_m(u0, t, nt = 80): #construct matrix M for a given time step and forward model
    # for more information, refer to adjoint_model
    u0 = forward_model(u0)
    u = np.zeros((nt, nz))
    u[:, :] = u0
    M = np.zeros((nz, nz))
    for i in range(0, nz-1):
        M[i, i] = 1+dt*(-(u[-t-1, i+1]+u[-t-1, i-1])/(2*dz) - 2*nu/(dz**2))
        M[i, i+1] = dt*(-u[-t-1, i]/(2*dz)+nu/(dz**2))
        M[i+1, i] = dt*(u[-t-1, i]/(2*dz)+nu/(dz**2))
    return M 

B_prev = generate_B_matrix(z)
Binv = np.linalg.inv(B_prev)
Q = np.zeros((len(z), len(z)))
Q[15, 15] = 0.05**2 #set the Gaussian error forecast matrix
Q[45, 45] = 0.1**2 # with variances
M = construct_m(x_b, 0)
temp = np.linalg.inv(R  + h[0, ...]@B@h[0, ...].T) #to ease calculation
P = (np.identity(len(z)) - B@h[0, ...].T@temp@h[0, ...])@B #analysis error covariance matrix 
xb = np.ones(61)
# compute first term
x_a_0 = forward_model(xb)[0]
x_b_n = M@x_a_0 + dt*10*np.sin(.375*np.pi*(0+0.5*dt))*np.exp(-100*z**2)

for i in range(-1, 78): #update, compute all of the remaining terms
    M = construct_m(x_b, i+1)
    B_new = np.array(M@P@M.T)  + Q #most expensive operation (mat mat mat mult)
    temp = np.linalg.inv(R  + h[0, ...]@B_prev@h[0, ...].T)
    x_a_n = x_b_n + B_new@h[0, ...].T@temp@(y[i+1]-h[0, ...]@x_b_n)
    P = (np.identity(len(z)) - B_prev@h[0, ...].T@temp@h[0, ...])@B_prev
    x_b_n = M@x_a_n +dt*10*np.sin(.375*np.pi*(i+2+0.5*dt))*np.exp(-100*z**2)
    B_prev = B_new
x_b_n+=x_b

##3D Var attempt
def J(x0):
    x = forward_model(x0)
    LS = (x0-x_b)@(Binv@(x0-x_b))
    x = forward_model(np.ones(61)) 
    final = np.array([])
    y_h_x = np.array([])
    for i in range(0, 80):
        temp = np.stack((x[i][15], x[i][45]))
        temp = y[i] - temp
        y_h_x = np.concatenate((y_h_x, temp), axis = None)
        temp = Rinv@temp
        final = np.concatenate((final, temp), axis = None)
    final = np.reshape(final, (-1, 2))
    y_h_x = np.reshape(y_h_x, (-1, 2))
    RS = y_h_x.T.flatten()@final.flatten()
    return LS - RS
```

```python
plt.figure(figsize=(16, 6))
plt.plot(z, np.ones(len(z)), label = 'Initial background' )
plt.plot(z, x_a, label = 'New $x_b$ (or $x_0$) / Final $x_a$ (Single OI)')
plt.plot(z, x_a_2, '--', label = 'New $x_b$ (or $x_0$) / Final $x_a$ (Multiple OI)')
plt.plot(z, x_b_n, '--', label = 'New $x_b$ (or $x_0$) / Final $x_a$ (Kalman Filter)')
plt.title('Optimal $z_0$ (or $z_b$) after optimisation / DA ($z_b = 1$)')
plt.ylabel('$z_0(z)$')
plt.xlabel('$z$')
plt.legend()
```
